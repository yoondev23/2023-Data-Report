```python
import folium
import branca
from folium.plugins import HeatMap
import pandas as pd
from collections import defaultdict
# Draw a basemap
```

```python
disable_data = pd.read_csv(' 2023 DATA/(장애인)지하철 최다이용정류장_경로추가.csv')
disable_data.head(10)
```

```python
#BASE MAp
disable_map = folium.Map(location=[35.1576059507137, 129.05917617141483],tiles ="CartoDB positron", zoom_start=11)


color_map = branca.colormap.LinearColormap(
    colors=['white','#DFAB94','#E3936F','#9D4B25','#884323'],
    vmin=0.1, vmax=1.0,
).to_step(n=15)

# color 분기 나누기
steps=100
gradient_map_m=defaultdict(dict)
for i in range(steps):
    gradient_map_m[1/steps*i] = color_map.rgb_hex_str(1/steps*i)
    
#color_map 이름 달기
color_map.caption='장애인 지하철 이용빈도'
disable_map.add_child(color_map)

# Add heatmap| gradient=gradient_map_m,
HeatMap(
    data = disable_data[['ycoord_2','xcoord_2']], 
    radius=12,
    gradient =  gradient_map_m,
).add_to(disable_map)

# Display the map
disable_map
```

```python
disable_map.save('C:/Users/ohkwa/Desktop/disable_map.html')
```
